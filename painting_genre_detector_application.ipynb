```python
labels = ['Abstractism', 'Baroque', 'Byzantine', 'Cubism', 'Expressionism', 'High_Renaissance',
             'Impressionism', 'Mannerism', 'Muralism', 'Northern_Renaissance', 'Pop_Art',
             'Post-Impressionism', 'Primitivism', 'Proto_Renaissance', 'Realism', 'Romaticism',
             'Suprematism', 'Surrealism', 'Symbolism']
```

```python
from tensorflow.keras.preprocessing import image
#loading trained model
CNN_model = tf.keras.models.load_model('/content/gdrive/My Drive/Data/models/my_model')

image_path = '/content/gdrive/My Drive/Data/demo_images/Pop_Art/Andy_Warhol_15.jpg'
input_image = image.load_img(image_path, target_size=(224, 224))
plt.imshow(input_image)
plt.show()
#preprocess the image
input_image = img_to_array(input_image)
input_image = input_image.reshape((1, input_image.shape[0], input_image.shape[1], input_image.shape[2]))
#this will apply VGG16 preprocessing which matches how we trained
input_image = preprocess_input(input_image)

prediction = CNN_model.predict(input_image)
# printing the prediction to show how the model outputs it's prediction using OneHotEncoder
print(prediction)
# find index of the highest argument(which genre the model believes the painting is) and printing the label of that index
index = np.argmax(prediction)
print(labels[index])
```
